```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
usa = gpd.read_file("../data/map/US_State_Boundaries.geojson")
```

```python
usa['geometry'][24]
```

```python
# plt.figure(figsize=(18, 12), dpi=150)
usa.plot(color="white", edgecolor="black", linewidth=0.5, figsize=(22,12))
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.xlim([-180, -65])
plt.ylim([22, 72])
plt.title("Rainy Cities in the U.S.A.")
plt.show()
```
